**Imports Libraries**

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Import Daata**

```python

# Import Data
df = pd.read_csv('https://raw.githubusercontent.com/jsdhami/Python-For-Research/main/Data/hourly_ktm_weather_data.csv')

df = pd.DataFrame({'date':df['date'], 'temp':df['temperature_2m'], 'ppt':df['precipitation'], 'windspeed':df['wind_speed_10m']})
df.head()
print(len(df))
```

```python

df = df.tail(25)
df.head()
# print all
print(df.to_string())
```

**Creating Data**

```python
x=df['date']
y=df['temp']
z=df['ppt']
w=df['windspeed']
```

**1. Line Ploting**

```python
#Line Ploting
# Creating Fig.
fig, ax = plt.subplots(figsize=(15, 5))

ax.plot(x, y,  label='Temperature', color='red')
ax.plot(x, z, label='precipitation', color='yellow')
ax.plot(x, w, label='wind speed', color='blue')

#label
ax.set_ylabel('Temperature,Precipitation,Wind Speed', color='violet')
ax.set_xlabel('Date')
ax.set_title('Weather Data  Of KTM 2024', color='orange')

# Set tick/Tick Label
ax.set_xticks(x[::1])
ax.set_xticklabels(x[::1], rotation=85, color='yellow')
ax.set_ylim(0, 40)

ax.legend()

plt.grid(False)

plt.show()
```

**2. Scatter Ploting**

```python
# Scatter Ploting
# Creating Fig,
fig, ax = plt.subplots(figsize=(15, 5))

ax.scatter(x, y,  label='Temperature',color='red')
ax.scatter(x, z, label='precipitation', color='yellow')
ax.scatter(x, w, label='wind speed', color='blue')

#label
ax.set_ylabel('Temperature,Precipitation,Wind Speed')
ax.set_xlabel('Date')
ax.set_title('Weather Data  Of KTM 2024')

# Stick label/Ticks Labels
ax.set_xticks(x[::1])
ax.set_xticklabels(x[::1], rotation=85)
ax.set_ylim(0, 40)

plt.legend()

plt.grid(False)

plt.show()
```

**10. Create One(2x2) Axes**

```python
# Create one (2x2) Axes
# Creating Figure
fig, ax = plt.subplots(2, 2)
fig.set_size_inches(20, 10)

#Temperature
ax[0, 0].plot(x, y, label='Temperature', color='red')

#label
ax[0, 0].set_ylabel('Temperature')
ax[0, 0].set_xlabel('Date')
ax[0, 0].set_title('Temperature of Ktm 2024')

# Sticks/Ticks label
ax[0, 0].set_xticks(x[::1])
ax[0, 0].set_xticklabels(x[::1], rotation=90, fontsize=7, color='white')

#Precipitation
ax[0, 1].plot(x, z, label='Precipitation', color='yellow')

#label
ax[0, 1].set_ylabel('Precipitation')
ax[0, 1].set_xlabel('Date')
ax[0, 1].set_title('Precipitation of Ktm 2024')

# Sticks/Ticks Labels
ax[0, 1].set_xticks(x[::1])
ax[0, 1].set_xticklabels(x[::1], rotation=90, fontsize=7, color='yellow')

#wind speed
ax[1, 0].plot(x, w, label='Wind Speed', color='blue')

#label
ax[1, 0].set_ylabel('Wind Speed')
ax[1, 0].set_xlabel('Date')
ax[1, 0].set_title('Wind Speed of Ktm 2024')

# Sticks/Ticks Labels
ax[1, 0].set_xticks(x[::1])
ax[1, 0].set_xticklabels(x[::1], rotation=90, fontsize=7, color='white')

#All Three
ax[1, 1].plot(x, y, label='Temperature', color='red')
ax[1, 1].plot(x, z, label='Precipitation', color='yellow')
ax[1, 1].plot(x, w, label='Wind Speed', color='blue')

#label
ax[1, 1].set_ylabel('Temperature, Precipitation, Wind Speed')
ax[1, 1].set_xlabel('Date')
ax[1, 1].set_title('Weather Data of KTM 2024')

# Sticks/Ticks Labels
ax[1, 1].set_xticks(x[::1])
ax[1, 1].set_xticklabels(x[::1], rotation=90, fontsize=7, color='yellow')

plt.tight_layout()

plt.legend()

plt.grid(False)

plt.show()



```
